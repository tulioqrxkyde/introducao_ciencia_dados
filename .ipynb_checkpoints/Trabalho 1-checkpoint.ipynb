### 1.4 Renomeando Colunas 

```python
new_names = ['Código Órgão', 'Código UASG', 'Modalidade de Licitação', 'Descrição Lei', 'Nº Inciso', 'Nº Processo', 
             'Quantidade Total do Item', 'Valor Estimado', 'Número Aviso',  'Objeto', 'Fundamento Legal', 'Justificativa',
             'Data Declaração', 'Responsável', 'Cargo Resp.', 'Ratificação', 'Nome Resp. Ratif.', 'Cargo Resp. Ratif.', 
             'Data Publicação']
```

```python
df.rename(columns=dict(zip(df.columns, new_names)), inplace=True)
```

### 1.6 Nulos!?

```python
df.loc[:, df.isnull().any()].columns
```

#### 1.6.1 Porcentagem de Nulos

```python
(df.isnull().sum() * 100/ len(df)).sort_values(ascending=False)
```

### 1.7 Removendo valores nulos

```python
df = df.loc[:,df.notna().any(axis=0)]
df
```

```python
values = round((df.isnull().sum() * 100/ len(df)),2).sort_values(ascending=True)
plt.plot(df.columns,values,label='launches per year')

plt.show()
```

### 1.8 Removendo Tags desnecessárias (Objeto:, Fundamento Legal: e Justificativa:)

```python
df = df.replace(to_replace=r'^\w*:|^\D*:', value='', regex=True)
df
```

### 1.9 Formatando datas para padrão DD/MM/YYYY

```python
df['Data Declaração'] = pd.to_datetime(df['Data Declaração'], errors='coerce').dt.strftime("%d/%m/%Y")
df['Data Publicação'] = pd.to_datetime(df['Data Publicação'], errors='coerce').dt.strftime("%d/%m/%Y")
df['Ratificação'] = pd.to_datetime(df['Ratificação'], errors='coerce').dt.strftime("%d/%m/%Y")
df
```
